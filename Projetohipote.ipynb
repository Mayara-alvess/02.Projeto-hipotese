```python
import pandas as pd
from scipy.stats import shapiro
from tabulate import tabulate

# Criar uma lista de dicionários para armazenar os resultados
resultados = []

# Definir as variáveis a serem testadas
variaveis = ['bpm', 'streams_clean', 'soma_playlist', 'in_spotify_charts',
             'in_apple_charts', 'in_deezer_charts', 'in_spotify_playlists',
             'in_apple_playlists', 'in_deezer_playlists', 'danceability__',
             'valence__', 'energy__', 'acousticness__', 'instrumentalness__',
             'liveness__', 'speechiness__']

# Iterar sobre cada variável e aplicar o teste de Shapiro-Wilk
for variavel in variaveis:
    stat, p_valor = shapiro(df[variavel].dropna())
    resultados.append({'Variável': variavel, 'Estatística de teste': stat, 'Valor p': p_valor})

# Criar um DataFrame a partir da lista de resultados
df_resultados = pd.DataFrame(resultados)

# Adicionar negrito aos nomes das variáveis
df_resultados['Variável'] = '\033[1m' + df_resultados['Variável'] + '\033[0m'

# Converter DataFrame para formato de tabela
tabela = tabulate(df_resultados, headers='keys', tablefmt='pipe', showindex=False)

# Imprimir tabela
print("\033[1mTabela Resumo dos Testes de Shapiro-Wilk\033[0m\n")
print(tabela)


```

```python

from scipy.stats import mannwhitneyu

##Resumir os dados em tabela##

# Lista de nomes das colunas
variaveis = ['danceability__', 'valence__', 'energy__', 'acousticness__', 'instrumentalness__', 'liveness__', 'speechiness__', 'bpm']

# Dicionário para armazenar os resultados do teste de Mann-Whitney U
resultados = {'Variável': variaveis, 'Estatística U': [], 'Valor p': []}

# Iterar sobre cada variável
for variavel in variaveis:
    # Filtrar os dados para 'Alto' e 'Baixo' para a variável atual e converter para array numpy
    Alto = df[df['categoria_streams_clean'] == 'Alto'][variavel].dropna()
    Baixo = df[df['categoria_streams_clean'] == 'Baixo'][variavel].dropna()

    # Executar o teste de Mann-Whitney U para a variável atual
    statistic, pvalue = mannwhitneyu(Alto, Baixo)

    # Adicionar os resultados ao dicionário
    resultados['Estatística U'].append(statistic)
    resultados['Valor p'].append(pvalue)

# Criar DataFrame com os resultados
df_resultados = pd.DataFrame(resultados)

# Estilizar o DataFrame para deixar apenas os nomes das variáveis em negrito
df_resultados_styled = df_resultados.style.apply(lambda x: ['font-weight: bold' if x.name == 'Variável' else '' for i in x])

# Adicionar um título à tabela de resumo
print("\033[1mTabela Resumo dos Testes de Mann-Whitney U\033[0m\n")

# Exibir DataFrame estilizado
df_resultados_styled

#H0 = o numero de streams entre as caracteristicas da música alto e baixo é o mesmo#
#H1 = o numero de streams entre as caracteristicas da música alto e baixo é diferente#

##valor p >0,05 Não existe correlação entre os streams e as categorias músicais##
##valor p <0,05 Existe  correlação entre streams e as categorias músicais##
```

```python
import numpy as np
from scipy.stats import ttest_ind

## resumir os dados em tabelas##

# Lista de nomes das colunas
variaveis = ['danceability__', 'valence__', 'energy__', 'acousticness__', 'instrumentalness__', 'liveness__', 'speechiness__', 'bpm']

# Dicionário para armazenar os resultados do teste t
resultados = {'Variável': variaveis, 't_valor': [], 'p_valor': []}

# Iterar sobre cada variável
for variavel in variaveis:
    # Filtrar os dados para 'Alto' e 'Baixo' para a variável atual e converter para array numpy
    Alto = df[df['categoria_streams_clean'] == 'Alto'][variavel].dropna()
    Baixo = df[df['categoria_streams_clean'] == 'Baixo'][variavel].dropna()

    # Executar o teste t para a variável atual
    resultado_teste = ttest_ind(Alto, Baixo)

    # Adicionar os resultados ao dicionário
    resultados['t_valor'].append(resultado_teste.statistic)
    resultados['p_valor'].append(resultado_teste.pvalue)

# Criar DataFrame com os resultados
df_resultados = pd.DataFrame(resultados)

# Estilizar o DataFrame para deixar apenas os nomes das variáveis em negrito
df_resultados_styled = df_resultados.style.apply(lambda x: ['font-weight: bold' if x.name == 'Variável' else '' for i in x])

# Adicionar um título à tabela de resumo
print("\033[1mTabela Resumo dos Testes T \033[0m\n")

# Exibir DataFrame estilizado
df_resultados_styled


```

```python

##Calculado, count, mean, max, min, std, var##

# Agrupar o DataFrame pela coluna 'categoria_streams_clean'
grupo = df.groupby('categoria_streams_clean')

# Calcular as estatísticas resumidas para cada variável
estatisticas = grupo.agg({'danceability__': ['count', 'mean', 'max', 'min', 'std','var'],
                           'valence__': ['count', 'mean', 'max', 'min', 'std','var'],
                           'energy__': ['count', 'mean', 'max', 'min', 'std','var'],
                           'acousticness__': ['count', 'mean', 'max', 'min', 'std','var'],
                           'instrumentalness__': ['count', 'mean', 'max', 'min', 'std','var'],
                           'liveness__': ['count', 'mean', 'max', 'min', 'std','var'],
                           'speechiness__': ['count', 'mean', 'max', 'min', 'std','var'],
                           'bpm': ['count', 'mean', 'max', 'min', 'std','var']})

# Transpor o DataFrame para exibir as estatísticas resumidas uma abaixo da outra
estatisticas_transposto = estatisticas.T

# Converter o DataFrame transposto em uma string formatada
tabela_formatada = estatisticas_transposto.to_string()

# Mostrar a tabela formatada
print(tabela_formatada)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Selecionar as variáveis para plotagem
variaveis = ['danceability__', 'energy__', 'valence__', 'acousticness__', 'liveness__', 'speechiness__', 'instrumentalness__']

# Configurar a largura das barras
largura_barra = 0.45
indice = np.arange(len(variaveis))

# Inicializar listas para armazenar as médias de streams para cada variável
medias_streams_alto = []
medias_streams_baixo = []

# Calcular as médias de streams para "Alto" e "Baixo" para cada variável
for variavel in variaveis:
    medias_streams_baixo.append(estatisticas[variavel].loc['Baixo', 'mean'])
    medias_streams_alto.append(estatisticas[variavel].loc['Alto', 'mean'])

# Criar o gráfico de barras
plt.figure(figsize=(12, 4))  # Reduzindo a altura do gráfico

barra_baixo = plt.bar(indice - largura_barra/2, medias_streams_baixo, width=largura_barra, label='Baixo', color='blue')
barra_alto = plt.bar(indice + largura_barra/2, medias_streams_alto, width=largura_barra, label='Alto', color='green')

# Adicionar rótulos com porcentagem nas barras
for barra in barra_baixo + barra_alto:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2., altura, '%.2f' % (altura), ha='center', va='bottom')

plt.xlabel('Variável')
plt.ylabel('Média de Streams')
plt.title('Médias de Streams para as Categorias Alto e Baixo')
plt.xticks(indice, variaveis)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
X = df['soma_playlist'].values
Y = df['streams_clean'].values
```

```python
from scipy.stats import pearsonr

# Calculando o coeficiente de correlação #
r, _ = pearsonr(X, Y)
print(f'Coeficiente de correlação: {r}')
```

```python

from scipy.stats import pearsonr, linregress

##Calculando regressão##

# Scatter plot
plt.scatter(X, Y, color='gray', label='Dados')

# Calcular a linha de tendência usando regressão linear
slope, intercept, _, _, _ = linregress(X, Y)
trendline = intercept + slope * X

# Plotar a linha de tendência
plt.plot(X, trendline, color='green', label='Linha de Tendência')

# Exibir o gráfico
plt.xlabel('Variável Independente')
plt.ylabel('Variável Dependente')
plt.title('Scatter plot com Linha de Tendência')
plt.legend()
plt.show()

# Calcular e exibir o coeficiente de correlação e o valor p
r, p_value = pearsonr(X, Y)
print(f'Coeficiente de correlação: {r}')
print(f'Valor p: {p_value}')
```
